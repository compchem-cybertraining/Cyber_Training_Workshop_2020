## Ground-state molecular dynamics simulation of Polyphenylene Ethynylene Dendrimer (PPE)

A ground state molecular dynamics simulation was performed following the procedure from:
* <i>J. Chem. Theory Comput. 2020, 16, 5771-5783.</i> 

Important parameters include:
* `grms_tol` = 0.0001
* `verbosity` = 5
* `natoms` = 48
* `rnd_seed` = 1,646,777
* `n_class_steps` = 13000
* `decoher_type` = 0
* `dotrivial` = 0
* `quant_step_reduction_factor` = 1d0
* `verbosity` = 2 (in the outpu & log parameters section)

It is also important to note that the velocity of each atom were set to zero: ($v_x = v_y = v_z = 0$).


```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
time = np.loadtxt('time.txt', comments="#", skiprows=0, delimiter='\t')
temp = np.loadtxt('temp.txt', comments="#", skiprows=0, delimiter='\t')
```

```python
# plot the data
plt.plot(time, temp,"b--", label="Temp")  
plt.xlabel("time, fs")
plt.ylabel("Temperature, K")
plt.title('Ground state molecular dynamics simulation')
plt.legend()
plt.show()

#print(time.shape)
```

```python
# load energetic data
time_energy = np.loadtxt('E_time.txt', comments="#", skiprows=0, delimiter='\t')
kinetic = np.loadtxt('kinetic.txt', comments="#", skiprows=0, delimiter='\t')
potential = np.loadtxt('potential.txt', comments="#", skiprows=0, delimiter='\t')
total_E = np.loadtxt('totalE.txt', comments="#", skiprows=0, delimiter='\t')
```

```python
# make the plots
plt.plot(time_energy, kinetic,"k--", label="KE")  
plt.plot(time_energy, potential,"b--", label="PE") 
plt.plot(time_energy, total_E,"g-", label="KE + PE") 
plt.xlabel("time, fs")
plt.ylabel("Energy, eV")
plt.title('Ground state molecular dynamics simulation')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
