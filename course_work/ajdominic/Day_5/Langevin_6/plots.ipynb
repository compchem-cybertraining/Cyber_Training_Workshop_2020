```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
time = np.loadtxt('time.out', comments="#", skiprows=0, delimiter='\t')
temp = np.loadtxt('temp.out', comments="#", skiprows=0, delimiter='\t')
```

```python
# plot the data
plt.plot(time, temp,"b--", label="Temp")  
plt.xlabel("time, fs")
plt.ylabel("Temperature, K")
plt.title('Ground state molecular dynamics simulation')
plt.legend()
plt.show()

#print(time.shape)
```

```python
# load energetic data
time_energy = np.loadtxt('E_time.txt', comments="##", skiprows=1, delimiter='\t')
kinetic = np.loadtxt('kinetic.txt', comments="##", skiprows=1, delimiter='\t')
potential = np.loadtxt('potential.txt', comments="##", skiprows=1, delimiter='\t')
total_E = np.loadtxt('total_E.txt', comments="##", skiprows=1, delimiter='\t')
```

```python
# make the plots
plt.plot(time_energy, kinetic,"k--", label="KE")  
plt.plot(time_energy, potential,"b--", label="PE") 
plt.plot(time_energy, total_E,"g-", label="KE + PE") 
plt.xlabel("time, fs")
plt.ylabel("Energy, eV")
plt.title('Ground state molecular dynamics simulation')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
